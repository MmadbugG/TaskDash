```python
def float_list(list):
    return [float(i) for i in list]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = []
with open('exp_fo.txt', 'r') as f:
    while(True):
        line = f.readline()
        if '-' in line:
            break
    for i in range(4):
        x.append(float_list(f.readline().split()))
```

```python
for i in range(1,4):
    plt.plot(x[0], x[i])
```
